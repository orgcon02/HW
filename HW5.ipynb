```python
import requests
import io
import pandas
```

```python
stock='TLSA'
date='2016-01-01'
def stock(stock,date):
    url = "http://finance.google.com/finance/historical"
    params = {}
    params['q'] = stock
    params['startdate'] = date
    params['output'] = 'csv'
    
    r = requests.get(url, params=params)
    r=pandas.read_csv(io.StringIO(r.text))

    return r
```

```python
stock(stock,date)
```

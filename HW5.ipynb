```python
import requests
```

```python
def stock(N,D):
    url = "http://finance.google.com/finance/historical"
    params = {}
    params['q'] = N
    params['startdate'] = D
    params['output'] = 'csv'
    
    r = requests.get(url, params=params)

    print(r.text)
```

```python
stock('TSLA','2016-01-01')
```
